```python
import numpy as np

# Generate a 10x5 array with random integers between 1 and 100
data1 = np.random.randint(1, 101, size=(10, 5))

# Generate a 5x5 array with random integers between 1 and 18
data2 = np.random.randint(1, 18, size=(10, 5))

# Concatenate data1 and data2 horizontally
combined_data = np.hstack((data1, data2))

print("Generated Data1:\n", data1)
print("\nGenerated Data2:\n", data2)
print("\nCombined Data Horizontally:\n", combined_data)
sort = np.sort(combined_data)
print("\nSort Combined Data:\n", np.sort(sort))

#SEARCHING
s = np.where(sort==4)
print(s)

```
